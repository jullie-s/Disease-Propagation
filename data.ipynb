### Adding neccassary packages

```python
import os
from sodapy import Socrata
from dotenv import load_dotenv
import pandas as pd
import json
```

### Gather Token Info

```python
load_dotenv()
token = os.getenv("APP_TOKEN")
```

### App Credentials

```python
domain = "data.cdc.gov"
identifier = "n8mc-b4w4"
```

### Gathering Data

```python
client = Socrata(domain, token)
results = client.get(identifier)
df = pd.DataFrame(results)
```
